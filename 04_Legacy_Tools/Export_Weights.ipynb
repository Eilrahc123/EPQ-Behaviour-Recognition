```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# 1. 严格按照训练时的结构手动搭建模型
model = Sequential()
model.add(LSTM(64, return_sequences=True, activation='tanh', input_shape=(60, 258)))
model.add(LSTM(128, return_sequences=True, activation='tanh'))
model.add(LSTM(64, return_sequences=False, activation='tanh'))
model.add(Dense(64, activation='tanh'))
model.add(Dense(32, activation='tanh'))
model.add(Dense(4, activation='softmax')) # 4 是你的 actions 类别数

# 2. 加载你原来的完整 H5 模型文件
# 假设你的文件名是 V4.5_student_behavior_lstm.h5
model.load_weights('V4.5_student_behavior_lstm.h5') 

# 3. 只导出权重部分（关键：使用 .weights.h5 后缀）
model.save_weights('V4.5_pure_weights.weights.h5')
print("✅ 权重已成功提取为 V4.5_pure_weights.weights.h5")
```

```python
import numpy as np
import tensorflow as tf

# 1. 加载你原本的 H5 模型
old_model = tf.keras.models.load_model('V4.5_student_behavior_lstm.h5', compile=False)

# 2. 提取所有权重并存为 numpy 压缩包
weights_list = old_model.get_weights()
np.savez('v45_weights_raw.npz', *weights_list)

print(f"✅ 成功将 {len(weights_list)} 组权重参数导出为 v45_weights_raw.npz")
```
